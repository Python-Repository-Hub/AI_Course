# Sum-up

```python
#################### hyper parameter setting ####################
learning_rate=1e-1
epochs = 1000

h = 10 ### number of hidden layer nodes
act = nn.Tanh() ### activation function
#################################################################

# prepare data
X_train = torch.tensor(_X_train, dtype = torch.float)
label = torch.tensor(_y_train, dtype = torch.float)
X_test = torch.tensor(_X_test, dtype = torch.float)
real = torch.tensor(_y_test, dtype = torch.float)

# prepare model
model = nn.Sequential(
        nn.Linear(1, h),
        act,
        nn.Linear(h, 1),
        )

# prepare optimization and loss function
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.1)

loss_array = []

# train the model
for i in range(epochs):
    optimizer.zero_grad()
    output = model(X_train)
    
    loss = loss_function(output, label)
    loss.backward() # back propagation
    optimizer.step() # update weight
    
    loss_array.append(loss)
    
    
prediction = model(X_test)
test_loss = loss_function(prediction, real)
print('test_loss:', test_loss.item())


# for plot
X = torch.tensor(_X, dtype = torch.float)
X_whole_pred = model(X).detach().numpy() 
prediction = prediction.detach().numpy()
real = real.detach().numpy()

# plot loss
plt.figure()
plt.plot(loss_array)
plt.axhline(test_loss.detach().numpy(), color='r', linestyle=':')
plt.show()

# plot model
plt.figure()
plt.scatter(_X_train, _y_train, label='train data')
plt.plot(X, X_whole_pred, label='model')
plt.title('train data')
plt.legend()
plt.show()

# plot inference
plt.figure()
plt.plot(X, X_whole_pred, label='model')
plt.scatter(X_test, real, label='real')
plt.scatter(X_test, prediction, label='prediction')
plt.title('test data')
plt.legend()
plt.show()
```
